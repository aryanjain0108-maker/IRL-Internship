```python
#1. User Login Check
username="admin"
password="1234"
u=input("Enter the username:")
p=input("Enter the password:")
if(u==username and p==password):
    print("Login successful")
else:
    print("Invalid credentials")
```

```python
#2. Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
p=0
f=0
for i in marks:
    if i>=50:
        p=p+1
    else:
        f=f+1
print("The number of students passed: ",p)
print("The number of students failed: ",f)
```

```python
#3.Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
j=0
for i in names:
    names[j] = (i.lower()).strip()
    j =  j+1
print(names)
```

```python
#4.Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
    n = len(i)
    print("The length of this message '",i,"' is ", n)
    if n>10:
        print("It contains more than 10 characters")
    print("\n")    
```

```python
#5.Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
e=0
for i in logs:
    if i=="ERROR":
        e=e+1
print('The error count in logs is :',e)
```
