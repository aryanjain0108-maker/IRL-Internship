```python
# Social Media – Post Engagement Analyzer
def analyze_engagement(likes_list):
    total_likes = 0
    for likes in likes_list:
        total_likes += likes
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    return total_likes, status
likes_data = [200, 150, 300, 400]
total, post_status = analyze_engagement(likes_data)
print("Total Likes:", total)
print("Post Status:", post_status)
```

```python
#Healthcare – Medicine Stock Alert System
def monitor_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    return stock, status
medicine_stock = 6
stock_value, result = monitor_stock(medicine_stock)
print("Medicine Stock:", stock_value)
print("Status:", result)
```

```python
# Agriculture – Rainfall Adequacy Checker
def analyze_rainfall(rainfall_data, required_level):
    total_rainfall = 0
    for rain in rainfall_data:
        total_rainfall += rain
    average_rainfall = total_rainfall / len(rainfall_data)
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    return average_rainfall, status
rainfall_list = [60, 75, 80, 70, 75]
required = 70
average, rainfall_status = analyze_rainfall(rainfall_list, required)
print("Average Rainfall:", int(average))
print("Rainfall Status:", rainfall_status)
```

```python
#Social Media – Duplicate Account Detection
def detect_duplicates(usernames):
    unique_users = set()
    duplicate_found = False
    for user in usernames:
        if user in unique_users:
            duplicate_found = True
            break
        unique_users.add(user)
    if duplicate_found:
        return "Yes"
    else:
        return "No"
username_list = ["aryan", "rahul", "sneha", "aryan", "kiran"]
result = detect_duplicates(username_list)
print("Duplicate Accounts Found:", result)
```

```python
# Agriculture – Premium Crop Price Filter
def filter_premium_crops(prices):
    premium_list = []
    for price in prices:
        if price > 2000:
            premium_list.append(price)
    return premium_list
crop_prices = [1500, 2500, 1800, 3200, 2000]
premium_crops = filter_premium_crops(crop_prices)
print("Premium Crops:", premium_crops)
```

```python
#System Monitoring – Application Health Checker
def evaluate_system_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    return error_count, status
errors = 7
count, system_status = evaluate_system_health(errors)
print("Error Count:", count)
print("System Status:", system_status)
```

```python
#E-Learning – Student Attendance Eligibility System
def check_exam_eligibility(attendance_list):
    present_days = 0
    for status in attendance_list:
        if status == "P":
            present_days += 1
    attendance_percentage = (present_days / len(attendance_list)) * 100
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    return attendance_percentage, eligibility
attendance = ['P','P','A','P','P']
percentage, result = check_exam_eligibility(attendance)
print("Attendance Percentage:", percentage)
print("Exam Eligibility:", result)
```

```python
#Smart Electricity Bill Analyzer
def calculate_electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    return bill, status
units_consumed = int(input("Enter the units:"))
total_bill, usage_status = calculate_electricity_bill(units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
