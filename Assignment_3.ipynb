```python
#Employee Performance Bonus Eligibility
employees= {}
n = int(input("How many employees? "))
for i in range(n):
    name = input("Enter employee name: ")
    performance = input("Enter performance: ")
    employees[name] = performance
def top_performers(employees):
    max_score = max(employees.values())
    top_employees = [name for name, score in employees.items() if score == max_score]
    print(f"Top Performers Eligible for Bonus: {', '.join(top_employees)} (Score: {max_score})")
top_performers(employees)
```

```python
#Search Query Keyword Analysis
import string
def analyze_search_query(query):
    query = query.lower()
    query = query.translate(str.maketrans('', '', string.punctuation))
    words = query.split()
    word_count = {}
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1
    repeated_words = {word: count for word, count in word_count.items() if count > 1}
    return repeated_words
query = input("Enter a sentence:")
result = analyze_search_query(query)
print(result)
```

```python
#Sensor Data Validation
def valid_sensor_readings(sensor_readings):
    valid_readings = []
    for index, value in enumerate(sensor_readings):
        if value % 2 == 0:   
            valid_readings.append((index, value))
    return valid_readings
sensor_readings = []
n=int(input("Enter the numbers you wish to insert:"))
for i in range(n):
    i=int(input("Enter a number:"))
    sensor_readings.append(i)
result = valid_sensor_readings(sensor_readings)
print("Valid Sensor Readings (Hour, Value):")
print(result)
```

```python
#Email Domain Usage Analysis
def analyze_email_domains(emails):
    domain_count = {}
    for email in emails:
        domain = email.split("@")[1]
        domain_count[domain] = domain_count.get(domain, 0) + 1
    total_users = len(emails)
    for domain, count in domain_count.items():
        percentage = (count / total_users) * 100
        print(f"{domain}: {int(percentage)}%")
emails = ["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]
analyze_email_domains(emails)
```

```python
#Sales Spike Detection
def detect_sales_spikes(sales):
    average_sales = sum(sales) / len(sales)
    threshold = average_sales * 1.29
    for index, value in enumerate(sales):
        if value > threshold:
            print(f"Day {index + 1}: {value}")
sales = []
n=int(input("Enter the no. of sales:"))
for i in range(n):
    i=int(input("Enter the sales amount:"))
    sales.append(i)
detect_sales_spikes(sales)
```

```python
#Duplicate User ID Detection
def find_duplicate_users(user_ids):
    user_count = {}
    for user in user_ids:
        user_count[user] = user_count.get(user, 0) + 1
    for user, count in user_count.items():
        if count > 1:
            print(f"{user} → {count} times")
user_ids = []
n=int(input("Enter the number of users:"))
for i in range(n):
    i=input("Enter userID:")
    user_ids.append(i)
find_duplicate_users(user_ids)
```
