```python
#Mobile Recharge Validation System
def mobile_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans: 199, 299, 399, 599")
        return False
    return True
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        if mobile_recharge(amount):
            print("Recharge is successful")
            break
        else:
            print("Please try again.\n")
    except ValueError:
        print("Please enter a valid numeric amount.\n")
```

```python
#Inventory Reorder Alert System
def check_stock(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f"{product}:Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}:Stock OK (Stock = {quantity})")
products = {"Rice": 10,"Wheat": 25,"Sugar": 12,"Oil": 30,"Salt": 18}
check_stock(products)
```

```python
#Student Result Processing System
def evaluate_result(marks):
    total = 0
    for mark in marks:
        total += mark
    average = total / len(marks)
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"
student_marks = []
n=int(input("Enter the number of subjects:"))
for i in range(n):
    i=int(input("Enter the marks:"))
    student_marks.append(i)
avg, result = evaluate_result(student_marks)
print("Average Marks:", avg)
print("Result:", result)
```

```python
#Cab Fare Estimator with Retry Option
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    fare = base_fare + (distance * per_km_rate)
    if peak:
        fare = fare + (0.25 * fare)
    return fare
while True:
    try:
        distance = float(input("Enter distance travelled (in km): "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()
        peak = True if peak_input == "yes" else False
        total_fare = calculate_fare(distance, peak)
        print("Total Fare: ₹", total_fare)
    except ValueError:
        print("Please enter valid numeric distance.")
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        print("Thank you for using Cab Fare Calculator ")
        break
```

```python
#Employee Attendance Eligibility Checker
def check_eligibility(attendance):
    present_days = 0
    for status in attendance:
        if status == "P":
            present_days += 1
    percentage = (present_days / len(attendance)) * 100
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"
attendance_list = []
n=int(input("Enter the number of days:"))
for i in range(n):
    i=input("Enter the attendance(P/A):")
    attendance_list.append(i)
percent, result = check_eligibility(attendance_list)
print("Attendance Percentage:", percent, "%")
print("Status:", result)
```

```python
#Password Strength Checker
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character @#$)"
pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```
