```python
#Smart Parking Lot Management System
def smart_parking_system(capacity, vehicle_logs):
    current_vehicles = 0
    peak_usage = 0
    for log in vehicle_logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles
    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_usage)
    if current_vehicles > capacity:
        print("Parking Status: Exceeded Capacity! ALERT!")
    else:
        print("Parking Status: Available")
capacity=int(input("Parking Capacity:"))
vehicle_logs=[]
n=int(input("Enter number of vehicle logs:"))
for i in range(n):
    i=input("Vehicle logs:")
    vehicle_logs.append(i)
smart_parking_system(capacity,vehicle_logs)
```

```python
#Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5
    delay = 0
    if traffic.lower() == "high":
        delay += 10
    elif traffic.lower() == "medium":
        delay += 5
    if weather.lower() == "rainy":
        delay += 5
    elif weather.lower() == "stormy":
        delay += 15
    final_eta = base_time + delay
    return final_eta
distance_km = int(input("Distance(km):"))
traffic_level = input("Traffic Level:")
weather_condition = input("Weather:")
eta = estimate_delivery_time(distance_km, traffic_level, weather_condition)
print("Estimated Delivery Time:", eta, "minutes")
```

```python
#Movie Theatre Seat Occupancy Analyzer
def analyze_booking(total_seats, booked_seats):
    booked_count = 0
    for seat in booked_seats:
        if seat == 1:
            booked_count += 1
    occupancy = (booked_count / total_seats) * 100
    if occupancy == 100:
        status = "Housefull"
        suggestion = "Open Additional Show"
    elif occupancy >= 75:
        status = "Almost Full"
        suggestion = "Monitor Booking Trend"
    else:
        status = "Seats Available"
        suggestion = "No Additional Show Needed"
    return occupancy, status, suggestion
total_seats = int(input("Enter the number of seats:"))
booked_seats = [1] * 150
occupancy_percent, show_status, suggestion = analyze_booking(total_seats, booked_seats)
print("Occupancy:", int(occupancy_percent), "%")
print("Show Status:", show_status)
```

```python
#Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    average_cpu = total / len(cpu_readings)
    if average_cpu < 50:
        status = "Normal"
    elif average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"
    return average_cpu, status
cpu_data = []
n=int(input("Enter the number of cpu readings:"))
for i in range(n):
    i=int(input("Enter the cpu readings:"))
    cpu_data.append(i)
average, server_status = classify_server_load(cpu_data)
print("Average CPU Load:", int(average), "%")
print("Server Status:", server_status)
```

```python
#Smart Classroom Resource Usage Monitor
def monitor_resources(resource_usage):
    overused = []
    threshold = 8
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)
    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"
    return overused, alert
usage_data = {"Projector": 6,"AC": 9,"Lights": 4}
overused_resources, energy_alert = monitor_resources(usage_data)
print("Overused Resources:", ", ".join(overused_resources))
print("Energy Alert:", energy_alert)
```

```python
#Online Event Registration Capacity Controller
def manage_registrations(capacity, total_registrations):
    confirmed = 0
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
    waitlisted = total_registrations - confirmed
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"
    return confirmed, waitlisted, status
event_capacity = int(input("Enter the event capacity:"))
registrations = int(input("Enter the number of registrations:"))
confirmed_count, waitlist_count, registration_status = manage_registrations(event_capacity, registrations)
print("Confirmed Registrations:", confirmed_count)
print("Waitlisted Users:", waitlist_count)
print("Registration Status:", registration_status)
```
